```python
input_file = open("seq_sample.fastq","r")
input_list = input_file.readlines()
file_length = len(input_list)
input_file.close()
#read the file and store into a list

for seq in range(3, file_length, 8):
    #print (seq, seq+4) to check if we get quality score lines
    quality1 = input_list[seq].rstrip()
    #quality score lines of the first sequence in a pair
    #print(quality1)
    quality2 = input_list[seq+4].rstrip()
    #print (quality2)
    #quality score lines of the second sequence in a pair
    
    list1 = []
    for x1 in quality1:
        #print(x1, ord(x1))
        list1.append(ord(x1))
        #create a blank list and append each quality score into the list
    #print(list1) 
    #check
    total1 = 0
    for q_s1 in range(0,len(list1)):
        total1 = total1 + list1[q_s1]
        #add elements in list1 together
        average1 = total1 / len(list1)
        # the average Q score equals to sum divided by the number of elements in the list
    #print(total1)
    #print(len(list1))
    #print(average1)

    sample1 = input_list[seq-3]
    #print(sample1)
    sequence1 = input_list[seq-2]
    #print(sequence1)
    # the first sequence in a pair
    if average1 - 64 < 30:
    # check if Q Score <30, if yes:
        #print(sample1)
        #print(sequence1)
        #print out the sample number and sequence to check
        sample1_crs = input_list[seq+1]
        sequence1_crs = input_list[seq+2]
        #the corresponding sample number and sequence in that pair
        #print(sample1_crs)
        #print(sequence1_crs)
        #check
    
    #Here I want to do the same for quality2 as quality1
    list2 = []
    for x2 in quality2:
        #print(x2, ord(x2))
        list2.append(ord(x2))
    #print(list2)
    #check
    total2 = 0
    for q_s2 in range (0, len(list2)):
        total2 = total2 + list2[q_s2]
        average2 = total2 / len(list2)
    #print(total2)
    #print(len(list2))
    #print(average2)
    #check
    
    sample2 = input_list[seq+1]
    #print(sample2)
    sequence2 = input_list[seq+2]
    #print(sequence2)
    if average2 - 64 < 30:
        sample2_crs = input_list[seq-3]
        sequence2_crs = input_list[seq-2]
        #print(sample2_crs)
        #print(sequence2_crs)
        #crs is moved ahead because i want the first sequence in a pair to appear first
        #print(sample2)
        #print(sequence2)
        # print all sample numbers and sequence, including repeated sequences
        
    #now i want store all of these into a list to remove repeated sequences and to output
    list_below30 = []
    list_above30 = []
    if average1 - 64 < 30:
        #check if the first sequence in a pair has a Q Score below 30
        list_below30.append(sample1)
        #print(list_below30)
        list_below30.append(sequence1)
        #print(list_below30)
        list_below30.append(sample1_crs)
        #the sample number of the corresponding sequence to the sequence with Q Score below 30
        #print(list_below30)
        list_below30.append(sequence1_crs)
        #the corresponding sequence to the sequence with Q Score below 30
        #print(list_below30)
   
    if average2 - 64 < 30:
        #check if the second sequence in a pair has a Q Score below 30
        list_below30.append(sample2_crs)
        #print(list_below30)
        list_below30.append(sequence2_crs)
        #print(list_below30)
        list_below30.append(sample2)
        #print(list_below30)
        list_below30.append(sequence2)
        #print(list_below30)
        if len(list_below30) > 4:
            # some pairs have both sequences below 30 so i delete the repeated index
            #print(list_below30)
            del list_below30[0:4]
        #print(list_below30)
        out_file = open("below30.fastq","a")
        #create and open a file, use"a" to append elements one by one 
        for b30 in list_below30:
            #print(b30)
            #check
            out_file.write(b30)
            #go through elements in list_below30 and write into the file one by one
        out_file.close()
    
    if average1 - 64 >= 30:
        #check if the first sequence in a pair has a Q Score above 30
        sample1_crs = input_list[seq+1].rstrip()
        sequence1_crs = input_list[seq+2].rstrip()
        #define sample1_crs and sequence1_crs in this block
        list_above30.append(sample1)
        list_above30.append(sequence1)
        list_above30.append(sample1_crs)
        list_above30.append(sequence1_crs)
    
    if average2 - 64 >= 30:
        ##check if the second sequence in a pair has a Q Score above 30
        sample2_crs = input_list[seq-3]
        sequence2_crs = input_list[seq-2]
        #remove rstrip() so that they can be on different lines
        list_above30.append(sample2_crs)
        list_above30.append(sequence2_crs)
        list_above30.append(sample2)
        list_above30.append(sequence2)
        #print(list_above30)
        if len(list_above30) <= 4:
            #check for any leftover
            #print(list_above30)
            out_file = open("below30.fastq","a")
            #open the closed file and append
            for ab30 in list_above30:
                #print(ab30)
                out_file.write(ab30)
            out_file.close()
            #close the file
    if len(list_above30) > 4:
        #both sequences have to be above 30 and so only lists with length over 4 will pass to be written in the file
        #print(list_above30)
        del list_above30[0:4]
        #print(list_above30)
        out_file = open("above30.fastq","a")
        #create a new file and append
        for a30 in list_above30:
            #print(a30)
            out_file.write(a30)
        out_file.close()
        #close the file

print("Read pairs where both sequences with an average Q score above 30 have been stored in above30.fastq")
print("Read pairs where at least one of the sequences with an average Q score below 30 have been stored in below30.fastq")
#print out the result

print("The number of read pair copies stored depends on how many times you run this program.")
#Reminding that when ONE copy of the read pairs is stored when first time running it, and running the second time will have two copies.

#The coding is not concise because when I work through I check after every step is done so that each block is relatively independent, and there could be repetitive definition.    

        
```
